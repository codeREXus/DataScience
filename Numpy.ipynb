```python
import numpy as np

lst=[1,23,23,56,7]
arr=np.array(lst)
print(arr)
```

```python
arr1=np.arange(1,20,5)
print(arr1)
```

```python
arr2=np.array([[12,34,56,77],[77,65,43,21]])
print(arr2)
print(arr2.size)
# to print number of rows
print(np.size(arr2,0))
# to print number of columns
print(np.size(arr2,1))
# to get the shape directly 
print(arr2.shape)
```

```python
print(arr2.dtype)
```

```python
zero=np.zeros((3,5),int)
print(zero)
#simmilarly ones

# to transpose a matrix 
print(zero.T)
```

```python
# to get ones on diagonal elements use eye function
print(np.eye(2,dtype=int))
print(np.eye(7,dtype=int))
```

```python
# searchsorted0- returns the optimum location of a new element that should be entered at that specfic location
lst=np.array([1,23,4,65,45])
print(np.sort(lst,kind='quick'))
print(lst.searchsorted(242))
```

```python
np.random.seed(2)
arr1=np.random.randint(1,100,10)
print(arr1)
print(np.where(arr1%2==0,'even','odd'))
```

```python
np.random.seed(12)
arr1=np.random.randint(1,100,(4,4))
print(np.hsplit(arr1,2))
```
